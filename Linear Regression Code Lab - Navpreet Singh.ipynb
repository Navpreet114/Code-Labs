```python
import numpy as np
import matplotlib.pyplot as plt

#generating some random datapoints
np.random.seed(0)

#rand() gives you uniformly distributed pseudo random numbers
x = np.random.rand(50) * 10

print("x:", x)

#normal distribution = bell like curve
y = 2 * x + 1 + np.random.randn(50) * 2

print("y:", y)

#Use numpy's polyfit method to perform linear regression on given data (x, y)
#third argument which we set to '1', indicating that we want to fit a polynomial in the degree 1

slope, intercept = np.polyfit(x, y, 1)

#Plot data points and the regression line
plt.scatter(x, y, label="Data Points")
plt.plot(x, slope * x + intercept, color="red", label="Regression Line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

print("Slope:", slope)
print("Y-intercept:", intercept)
```

```python
#Generate a new x value for prediction
new_x = 7

m = slope
b = intercept

#predict the corresponding y value using the best fit line equation (y = mx + b)
predicted_y = m * new_x + b

print("For x =", new_x, "the predicted y-value is:", predicted_y)

#plot the origional data points and best fit line
plt.scatter(x, y, label="Data Points")
plt.plot(x, m * x + b, color="red", label="Best Fit Line")

#plot the predicted point
plt.scatter(new_x, predicted_y, color="green", label="Predicted Point")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

print("Slope:", slope)
print("Y-intercept:", intercept)
```
